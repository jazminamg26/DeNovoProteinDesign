### **Programming examples**

**Function**

```python
def count_cells(cells_per_field, number_of_fields):
    total = cells_per_field * number_of_fields
    return total

total_cells = count_cells(150, 8)
print(total_cells)
```

**Classes and Objects**

```python
class Bird:
    def __int__(self, name, diet, energy):
        self.name = name
        self.diet = diet
        self.energy = energy

    def feed(self):
        self.energy += 20
        print(f"{self.name} ate some {self.diet}. Its new energy is: {self.energy}")
```

```python
Donald = Bird("Donald", "Fruit", 200)
```
